```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('religious-demographics.csv')
df.head()
```

```python
df = df.replace("< 1%", "1%")

df.head()
```

```python
df.dtypes
```

```python
convert = lambda x: float(x.replace("%", ""))

for col in df.columns:
    if col != "State" and col != "Sample size":
        df[col] = df[col].apply(convert)
        
df.head()
```

```python
df.dtypes
```

```python
christian_cols = [2,3, 5, 8, 9, 11, 12]

df['Christian Total'] = df.iloc[:, [2,3, 5, 8, 9, 11, 12]].sum(axis=1)

df = df.iloc[:, [i for i in range(len(df.columns)) if i not in christian_cols]]

df.head()
```

```python
df = df.drop(["Jehovah's Witness", "Other Faiths", "Other World Religions", "Unaffiliated (religious \"nones\")", "Don't know"], axis=1)
df.head()
```

```python
df = df.drop("Sample size", axis=1)

df.head()
```

```python
df.to_csv("clean_religion_data.csv", index=False)
```
