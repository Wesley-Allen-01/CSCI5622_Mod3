```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dirty_data.csv')
df.head(67)
```

```python
df.iloc[:, 0].head(30)
```

```python
df = df.iloc[[1, 5, 6, 15], [0] + list(range(1, 613, 12))]
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
df.rename(columns={col : col.split("!!")[0].strip() for col in df.columns}, inplace=True)

df.head()
```

```python
df = df.T

df.head()
```

```python
df.columns = list(df.iloc[0])

df = df[1:]

df.reset_index(inplace=True)

# df.rename(columns={'index': 'State'}, inplace=True)

df.head()
```

```python
df.rename(columns={'index' : 'state'}, inplace=True)

df.head()
```

```python
df.columns = df.columns.str.replace(r'^\s+', '', regex=True)

# Display the updated column names
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
df = df.replace(',', ',', regex=True)

df.head()
```

```python
df.head()
```

```python
# Step 1: Remove commas from all columns except the first one (assuming it's non-numeric)
df.iloc[:, 1:] = df.iloc[:, 1:].replace(',', '', regex=True)

# Step 2: Convert columns to integers
df.iloc[:, 1:] = df.iloc[:, 1:].astype(int)


```

```python
df['total population'] = df.iloc[:, [1, 3]].sum(axis=1)


df.head(10)
```

```python
df['Population with Bachelor\'s degree'] = df.iloc[:, [2, 4]].sum(axis=1)
df.head()
```

```python
df['Percent With Degree'] = df['Population with Bachelor\'s degree'] / df['total population'] * 100

df.head()
```

```python
df.to_csv('clean_data.csv', index=False)
```
